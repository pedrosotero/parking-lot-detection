```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(rc={'figure.figsize':(15,8.27)})
```

# Curto

```python
curto_bof = pd.read_csv('dados/curto_BOF.csv', index_col=0)
curto_nn = pd.read_csv('dados/curto_NN.csv', index_col=0)

curto_bof['total'] = curto_bof['0'] + curto_bof['1'] + curto_bof['2'] + curto_bof['3'] + curto_bof['4'] + curto_bof['5']
curto_nn['total'] = curto_nn['0'] + curto_nn['1'] + curto_nn['2'] + curto_nn['3'] + curto_nn['4'] + curto_nn['5']

curto = pd.merge(curto_bof, curto_nn, how='inner', right_index=True, left_index=True, suffixes=['_bof', '_nn']).rename(columns={'total_bof': 'BoF', 'total_nn': 'NN'})
curto
```

```python
g = sns.lineplot(curto[['NN', 'BoF']])
g.set_title('Qtd. Vagas Ocupadas x Tempo (Curto)')
g.set_xlabel('Tempo (s)')
g.set_ylabel('Vagas ocupadas');
```

# Cortado

```python
cortado_bof = pd.read_csv('dados/cortado_BOF.csv', index_col=0)
cortado_nn = pd.read_csv('dados/cortado_NN.csv', index_col=0)

cortado_bof['total'] = cortado_bof['0'] + cortado_bof['1'] + cortado_bof['2'] + cortado_bof['3'] + cortado_bof['4'] + cortado_bof['5']
cortado_nn['total'] = cortado_nn['0'] + cortado_nn['1'] + cortado_nn['2'] + cortado_nn['3'] + cortado_nn['4'] + cortado_nn['5']

cortado = pd.merge(cortado_bof, cortado_nn, how='inner', right_index=True, left_index=True, suffixes=['_bof', '_nn']).rename(columns={'total_bof': 'BoF', 'total_nn': 'NN'})
cortado
```

```python
g = sns.lineplot(cortado[['NN', 'BoF']])
g.set_title('Qtd. Vagas Ocupadas x Tempo (Cortado)')
g.set_xlabel('Tempo (s)')
g.set_ylabel('Vagas ocupadas');
```
